```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
```

```python
from IPython.core.display import display_markdown
frase = 'Lorem Ipsum é simplesmente uma simulação de texto da indústria tipográfica e de impressos, e vem sendo utilizado desde o século XVI, quando um impressor desconhecido pegou uma bandeja de tipos e os embaralhou para fazer um livro de modelos de tipos. Lorem Ipsum sobreviveu não só a cinco séculos, como também ao salto para a editoração eletrônica, permanecendo essencialmente inalterado. Se popularizou na década de 60, quando a Letraset lançou decalques contendo passagens de Lorem Ipsum, e mais recentemente quando passou a ser integrado a softwares de editoração eletrônica como Aldus PageMaker.'
frase = frase + 'Ao contrário da crença popular, Lorem Ipsum não é simplesmente um texto aleatório. Tem raízes em uma peça da literatura clássica latina de 45 aC, com mais de 2.000 anos. Richard McClintock, um professor de latim no Hampden-Sydney College, na Virgínia, pesquisou uma das palavras latinas mais obscuras, consectetur, de uma passagem de Lorem Ipsum e, examinando as citações da palavra na literatura clássica...'

#configurações da nuvem
nuvem = WordCloud(
    background_color = 'white',
    width = 1000,
    height= 500
).generate(frase)
```

```python
#plotando o grafico
fig, ax = plt.subplots(figsize =(10,7))
ax.imshow(nuvem, interpolation = 'bilinear')
ax.set_axis_off() # tira as bordas
```

# MAPA DE CALOR

***Mapa de Calor*** <p>
O termo “mapa de calor” foi originalmente utilizado para identificar áreas com maior ou menor grau de temperatura ambiental. Serve para dimensionar a refrigeração, isolamento térmico e indicar o melhor posicionamento de pessoas e máquinas em um ambiente. Porém, como o termo entrega um entendimento mais fácil dos gráficos de frio e calor, passou a ser utilizado também em outras áreas.

Mapa de Calor.webp

```python
import pandas as pd
import plotly.express as px
```

```python
cidade = ['Canindé', 'Sobral', 'Fortaleza', 'Ipu', 'Ubajara', 'Alcantaras']
estado = ['Ceará', 'Ceará', 'Ceará', 'Ceará', 'Ceará', 'Ceará']

lat = [-4.3580, -3.6844, -3.7327, -4.3220, -3.8532, -3.5869 ]
longitude = [-39.3138, -40.3562, -38.5270, -38.5270, -40.9196, -40.5455 ]


Vendas = [100, 120, 90, 50, 250, 400]
```

```python
#dicionario

dic = {
    'Cidade' : cidade,
    'UF' : estado,
    'Lat' : lat,
    'Long' : longitude,
    'Vendas' : Vendas
}

dados = pd.DataFrame(dic)
dados
```

```python
# mapa de calor
px.density_mapbox(
    dados, 
    lat = 'Lat',
    lon = 'Long',
    z = 'Vendas',
    radius=30,
    center=dict(lat=-3.700, lon=-39.000),
    zoom=8,
    mapbox_style='stamen-terrain'
)
```
