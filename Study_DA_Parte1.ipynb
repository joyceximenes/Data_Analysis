# MATPLOTLIB

---



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**gerar vetor aleatorio de 1 a 100 com 10 amostras**

```python
np.random.seed(7)

y = np.random.randint(low = 1, high = 100, size = 10)

y
```

**grafico de linha**

```python
plt.plot(y)
```

**ajustando o grafico**

marker = marcador dos pontos

ms = tamanho do marcador

mec = cor da linha do marcador (k = preto)

ls = estilo da linha

plt.xlabel - escreve na horizontal

plt.ylabel - escreve na vertical

plt.title - adiciona o titulo

plt.grid - plota a linha de marcação



```python
plt.plot(y, color = '#749187', marker = 'o', ms = 5, mec = 'k',
          markerfacecolor = 'w', ls = '-.')
```

```python
plt.plot(y, color = '#749187', marker = 'o', ms = 5, mec = 'k',
          markerfacecolor = 'w', ls = '-.')

#segunda linha
plt.plot(y*2, marker = '+', color = 'k', ms = 5)

# rotulos
plt.xlabel("Eixo X", color = 'red', size = 12)
plt.ylabel('Eixo Y')
plt.title('Titulo', loc = 'center')

#gridlines
plt.grid(axis = 'y', color = 'green', linestyle = '--', linewidth = 1, alpha = 0.8)
```

**subplots**

**plt.subplot(1, 3, 1) - linha, colunas, posição**

**plt.tight_layout(pad = 4) - espaçamento**



```python
np.random.seed(6)
x = np.arange(1, 11) #valores espaçados dentro desse intervalo
y1 = np.random.randint(1, 50, 10)
y2 = np.random.randint(8, 78, 10)
y3 = np.random.randint(15, 130, 10)

plt.figure(figsize=(15,5))
plt.suptitle("Figura", fontsize = 10)

plt.subplot(1, 3, 1) # 1 linha, 3 colunas, 1a posição
plt.plot(x, y1, color = 'r')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 2) # 1 linha, 3 colunas, 1a posição
plt.plot(x, y2, color = 'g')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 3) # 1 linha, 3 colunas, 1a posição
plt.plot(x, y3, color = 'k')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad = 4)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15 , 5))
fig.suptitle('Figura')
ax[0].plot(x, y1, color = 'r')
ax[1].plot(x, y2, color = 'g')
ax[2].plot(x, y3, color = 'k')

for i in range(3):
  ax[i].set(title = f'Subplot {i+1}', xlabel = 'eixo x', ylabel = 'eixo y')
```

**scatter plot**

```python
figura = plt.figure()
ax = plt.axes()

ax.scatter(x, y, marker = 'o', color = 'r', label = 'data 1', alpha = 0.5)
ax.scatter(x*0.5, y*0.5, marker = 'v', color = 'k', label = 'data 2', alpha = 0.5)
ax.legend()
```

**dados economicos**
gapminder é atributo do plotly

```python
import plotly.express as px
px.data.gapminder()
```

**separando por atributo (pais)**

```python
tab = px.data.gapminder().query('country == "Brazil"').set_index("year")
display(tab)
```

```python
plt.plot(tab.index, tab['gdpPercap']) #parametros: ano e pib
plt.title("PIB per capita do Brasil")
plt.xlabel("PIB")
plt.ylabel('Tempo')
plt.show()
```

```python
title = 'Expectativa de vida x PIB percapita'

plt.figure(figsize=(10, 5))
plt.scatter(tab['lifeExp'], tab['gdpPercap'], cmap = 'viridis')
plt.xlabel('Expectativa')
plt.ylabel('Renda')
plt.title(title, loc = 'center')
```

**grafico de barra**

```python
plt.bar(x = tab.index, height = tab['pop'])
plt.title('População br')
plt.show()
```

# SEABORN

**seaborn: bibloteca baseada no matplotlib**

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

**definindo tema**

```python
sns.set_theme(style = 'darkgrid')
```

**carregando dados**

```python
datas = sns.load_dataset("tips")
datas.head()
```

**renomeando as colunas**

```python
datas.rename(columns = {
    'total_bill' : 'total conta', 
    'tip':'gorjeta',
    'sex':'sexo',
    'smoker':'fumante',
    'day':'dia',
    'time':'periodo',
    'size':'qtd de pessoas na mesa'}, inplace = True)

```

**grafico relplot**

```python
sns.relplot(x = 'total conta', y = 'gorjeta', data = datas)
```

**grafico com parametro de diferenciação**

```python
sns.relplot(x = 'total conta', y = 'gorjeta', data = datas, hue='sexo')
```

**estabelecendo um tipo e um parametro**

```python
sns.relplot(x = 'total conta', y = 'gorjeta', data = datas, kind = 'line', hue = 'fumante')
```

**grafico histplot**

```python
sns.histplot(data = datas, x = 'total conta')
```

```python
sns.histplot(data = datas, x = 'total conta', hue = 'fumante')
```

**grafico barplot - de barras**

```python
sns.barplot(data = datas, x = 'sexo', y = 'total conta', hue = 'fumante')
```

**grafico pairplot**

mostra as relações entre os atributos de forma geral

```python
sns.pairplot(datas)
```

```python
sns.pairplot(datas, hue = 'sexo')
```

**grafico boxsplot**

```python
sns.boxplot(data = datas, x = 'dia', y = 'total conta')
```

**grafico violinplot**

```python
sns.violinplot( data=datas, x='dia', y='total conta' );
```

**Gráfico swarmplot**

```python
sns.swarmplot(data=datas, x='total conta', y='sexo' );
```

**Gráfico jointplot**

```python
sns.jointplot( data=datas, x='total conta', y='gorjeta');
```
